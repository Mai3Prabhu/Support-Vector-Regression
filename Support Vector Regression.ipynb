```python
# Label encoding whenever we have two cataegories then we can replace the features with 0 and 1
# but when there are more than two categories then we use one hot encoding
```

```python
df.columns
```

```python
# independent and  dependent features
X = df[['tip', 'sex', 'smoker', 'day', 'time', 'size']]
y  = df['total_bill']
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
X_train
```

```python
# feature encoding
# the reason we need to divide this into train and test before feature encoding is to avoid data leakage as model should not have any knowledge
# of test data therefore we do train and test split and then  feature encoding.
```

```python
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()
```

```python
X_train['sex'] = le1.fit_transform(X_train['sex'])
X_train['smoker'] = le2.fit_transform(X_train['smoker'])
X_train['time'] = le3.fit_transform(X_train['time'])
```

```python
X_train.head()
```

```python
X_test['sex'] = le1.transform(X_test['sex'])
X_test['smoker'] = le2.transform(X_test['smoker'])
X_test['time'] = le3.transform(X_test['time'])
```

```python
X_test.head()
```

```python
# One hot encoding is applied alonng with column transformer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct=ColumnTransformer(transformers=[('onehot',OneHotEncoder(drop='first'),[3])],remainder='passthrough')
```

```python
import sys
import numpy
numpy.set_printoptions(threshold=sys.maxsize)


X_train = ct.fit_transform(X_train)
```

```python
X_test = ct.transform(X_test)
```

```python
X_test
```

```python
# Support Vector Regression (SVR)
from sklearn.svm import SVR
svr = SVR()
svr.fit(X_train,y_train)
```

```python
y_pred = svr.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
y_pred = svr.predict(X_test)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print("R-squared score:", r2)
print("Mean absolute error:", mae)
```

# Hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'kernel': [ 'rbf'],
    'C': [0.1, 1, 10],
    'gamma': ['scale', 'auto']
}
grid = GridSearchCV(SVR(), param_grid, refit=True)
grid.fit(X_train, y_train)
best_params = grid.best_params_
```

```python
print(best_params)
```

```python
grid_predictions = grid.predict(X_test)
r2_grid = r2_score(y_test, grid_predictions)
mae_grid = mean_absolute_error(y_test, grid_predictions)
print("R-squared score:", r2_grid)
print("Mean absolute error:", mae_grid)
#
```
